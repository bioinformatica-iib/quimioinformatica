```python
from rdkit import DataStructs
from rdkit import Chem
from rdkit.Chem.Fingerprints import FingerprintMols
from rdkit.Chem import Draw
ms = [
    Chem.MolFromSmiles('C1=CC=C(C=C1)C1=CC(=CC(=C1)C1=CC=CC=C1)C1=CC=CC=C1'),
    Chem.MolFromSmiles('C1=NC(=NC=N1)C1=NC(=NC(=N1)C1=NC=NC=N1)C1=NC=NC=N1'),
    Chem.MolFromSmiles('S1C=CC=C1'),
    Chem.MolFromSmiles('C1=CC=CC=C1')
    ]
fps = [FingerprintMols.FingerprintMol(x) for x in ms]
print(ms[0], ms[1], DataStructs.FingerprintSimilarity(fps[0],fps[1]))
```

```python
databasea = [
    Chem.MolFromSmiles("OC(O)C1=CSC(NC(=O)C2=CN=CC(Cl)=C2)=N1"), # query
    Chem.MolFromSmiles("NC(=S)C1=CC=C(NC(=S)CC2=CC(Br)=CO2)O1"), # I
    Chem.MolFromSmiles("C(C1CCCCN1CC1=CC=C2C=CC=CC2=C1)C1=CC=CC=C1"), # II
    Chem.MolFromSmiles("CCCCCCCCCCC#CC1=C(CCC(O)=O)C=CC=C1") # III
]
```

```python
Draw.MolsToGridImage(databasea)
```

```python
# fingerprints MACCS | old (similar to UNITY used in Flower DR 1998)
fps = [Chem.rdMolDescriptors.GetMACCSKeysFingerprint(x) for x in databasea]

print("query vs itself:", DataStructs.FingerprintSimilarity(fps[0],fps[0]))
print("query vs I:", DataStructs.FingerprintSimilarity(fps[0],fps[1]))
print("query vs II:", DataStructs.FingerprintSimilarity(fps[0],fps[2]))
print("query vs III:", DataStructs.FingerprintSimilarity(fps[0],fps[3]))
```
