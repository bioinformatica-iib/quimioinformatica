## Sistema Simplificado de Entrada Lineal Molecular: SMILES

SMILES significa "Sistema Simplificado de Entrada Lineal Molecular" y es una forma de representar moléculas como una cadena de caracteres.

Consideremos la molécula del etanol. La imagen a continuación muestra una representación que estamos acostumbrados a ver en química:

![ethanol](https://drive.google.com/uc?export=view&id=1pBnnNujVdkw43xpDOM27nzICgnn7EqJj)

Sin embargo, la representación SMILES de esta molécula sería `CCO`.

Podés leer más sobre SMILES en [este tutorial](https://archive.epa.gov/med/med_archive_03/web/html/smiles.html), aunque las reglas para átomos y enlaces también se resumen a continuación.

### Átomos

SMILES admite todos los elementos de la tabla periódica. Un átomo se representa usando su símbolo atómico correspondiente.  
Las letras mayúsculas se refieren a átomos no aromáticos; las letras minúsculas se refieren a átomos aromáticos.  
Si el símbolo atómico tiene más de una letra, la segunda debe ser minúscula.

### Enlaces

- `-` Enlace simple  
- `=` Enlace doble  
- `#` Enlace triple  
- `*` Enlace aromático  
- `.` Estructuras desconectadas  

Los enlaces simples son el valor predeterminado y, por lo tanto, no es necesario escribirlos.  
Por ejemplo, `CC` significa que hay un carbono no aromático unido a otro carbono no aromático por un enlace simple,  
y la computadora identificaría la estructura como el etano.  
También se asume que el enlace entre dos símbolos de átomos en minúscula es aromático.  
Un espacio en blanco finaliza la cadena SMILES.

### Ramificaciones

Una ramificación desde una cadena se especifica colocando el/los símbolo(s) SMILES de la ramificación entre paréntesis.  
Algunos ejemplos:

- `CC(O)C` → 2-Propanol  
- `CC(=O)C` → 2-Propanona  

### Anillos

Un anillo se especifica colocando un número directamente después del símbolo SMILES donde ocurre el cierre del anillo.  
Este número actúa como un marcador, indicando que los átomos con el mismo número están conectados, formando así un anillo.  
Por ejemplo:

- `C1CCCC1` → Ciclopentano  
- `n1ccccc1` → Piridina

# Parte 2: Softwares Quimioinformáticos

Existen bibliotecas de Python diseñadas específicamente para trabajar con datos químicos. Una de las bibliotecas más utilizadas en Python para ciencia de datos (o quimioinformática) se llama **RDKit**. RDKit es una biblioteca de quimioinformática de código abierto, desarrollada principalmente en C++ y en desarrollo continuo desde el año 2000. Nosotros utilizaremos la interfaz de Python de RDKit, aunque también existen interfaces en otros lenguajes.

![texto](https://www.rdkit.org/docs/_static/logo.png)

Fue desarrollado por Greg Landrum con numerosas contribuciones adicionales de la comunidad de código abierto RDKit. Tiene una interfaz de programación de aplicaciones para Python, Java, C++ y C#

- Homepage: http://www.rdkit.org Documentation, links
- Github (https://github.com/rdkit) Downloads, bug tracker, git repository
- Sourceforge (http://sourceforge.net/projects/rdkit) Mailing lists
- Blog (https://greglandrum.github.io/rdkit-blog/) Tips, tricks, random stuff
- Tutorials (https://github.com/rdkit/rdkit-tutorials) Jupyter-based tutorials for using the RDKit
- KNIME integration (https://github.com/rdkit/knime-rdkit) RDKit nodes for KNIME

RDKit proporciona un objeto de molécula que te permite manipular estructuras químicas. Cuenta con funcionalidades para leer y escribir distintos formatos de archivos moleculares, calcular propiedades moleculares y realizar búsquedas de subestructuras. Además, ofrece una amplia variedad de algoritmos de quimioinformática, como la generación de fingerprints moleculares, el cálculo de métricas de similitud y la computación de descriptores moleculares.

## Generar una molécula a partir de SMARTS

```python
# Define el patrón SMARTS que representa una estructura molecular
smarts = "C1C([Br])CCCC1"
```

```python
# Importar libreria de RDKit
from rdkit import Chem

# Convierte el patrón SMARTS en una molécula de RDKit
molecula = Chem.MolFromSmarts(smarts)
```

```python
# Imprime la molécula generada en formato SMARTS
molecula
```

### Actividad usando SMARTS
Edita el SMARTS para agregar la opción de que la molécula tenga un Bromo o un Cloro en la posición donde ahora hay sólo un Bromo.

Se puede generar el SMILES a partir del SMARTS usando la siguiente función:

```python
# Importar libreria de RDKit
from rdkit import Chem

# Obtener el SMILES de la molécula
smiles = (Chem.MolToSmiles(molecula))
```

```python
# Imprime la molécula generada en formato SMILES
smiles
```

## Sanitización de moléculas


Cuando se genera un objeto tipo `Mol` en RDKits, no solo se crea la molécula a partir del SMILES, sino que también verifica que la estructura sea válida en cuanto a la valencia de los átomos, los enlaces y otras reglas de química estructural.

Las funciones de análisis de moléculas, por defecto, realizan una operación de “sanitización” sobre las moléculas leídas. La idea es generar propiedades computadas útiles (como hibridación, pertenencia a anillos, etc.) para el resto del código y asegurarse de que las moléculas sean “razonables”: es decir, que puedan representarse con estructuras de Lewis completas con octetos.

Estos son los pasos involucrados, en orden:

- **`clearComputedProps`**: elimina cualquier propiedad computada que ya exista en la molécula, sus átomos y enlaces. Este paso siempre se realiza.

- **`cleanUp`**: estandariza un pequeño número de estados de valencia no estándar. Las operaciones de limpieza son:
  - Los nitrógenos neutros con valencia 5 y enlaces dobles a oxígenos se convierten a la forma zwitteriónica.
  - Los nitrógenos neutros con valencia 5 y enlaces triples a otro nitrógeno se convierten a la forma zwitteriónica.
  - Los fósforos neutros con valencia 5, con un doble enlace a oxígeno y otro a un carbono o a otro fósforo, se convierten a la forma zwitteriónica.
  - Los Cl, Br o I neutros, con exclusivamente oxígenos como vecinos y una valencia de 3, 5 o 7, se convierten a la forma zwitteriónica. Esto cubre compuestos como ácido cloroso, ácido clórico y ácido perclórico.

  Este paso no debería generar excepciones.

- **`cleanUpOrganometallics`**: estandariza un pequeño número de situaciones no estándar encontradas en compuestos organometálicos. Las operaciones de limpieza son:
  - Reemplaza enlaces simples desde átomos hipervalentes hacia metales por enlaces dativos.

  Este paso no debería generar excepciones.

- **`updatePropertyCache`**: calcula las valencias explícitas e implícitas de todos los átomos. Genera excepciones para átomos con estados de valencia superiores a los permitidos. Este paso siempre se realiza, pero si se “omite”, la prueba de valencias no estándar no se lleva a cabo.

- **`symmetrizeSSSR`**: ejecuta el algoritmo del conjunto más pequeño de anillos más pequeños simetrizado.

- **`Kekulize`**: convierte anillos aromáticos a su forma de Kekulé. Genera una excepción si un anillo no puede ser kekulizado o si se encuentran enlaces aromáticos fuera de los anillos.

- **`assignRadicals`**: determina el número de electrones radicales (si los hay) en cada átomo.

- **`setAromaticity`**: identifica los anillos y sistemas anulares aromáticos, marca como aromáticos a los átomos y enlaces, y ajusta los órdenes de enlace a aromático.

- **`setConjugation`**: identifica qué enlaces están conjugados.

- **`setHybridization`**: calcula el estado de hibridación de cada átomo.

- **`cleanupChirality`**: elimina etiquetas quirales de los átomos que no están hibridizados como sp³.

- **`adjustHs`**: agrega hidrógenos explícitos cuando es necesario para preservar la química. Esto suele ser necesario en heteroátomos de anillos aromáticos. El ejemplo clásico es el nitrógeno en el pirrol.

- **`updatePropertyCache`** (de nuevo): vuelve a calcular las valencias explícitas e implícitas de todos los átomos. Genera excepciones para átomos con estados de valencia superiores a los permitidos. Este paso es necesario para detectar algunos casos límite donde se aceptan átomos con valencias no físicas si están marcados como aromáticos.

Los pasos individuales se pueden activar o desactivar al llamar a `MolOps::sanitizeMol` o `Chem.SanitizeMol`.




Por defecto,  `sanitize=True` indica que RDKit intentará sanitizar la molécula después de construirla. La sanitización incluye varias verificaciones y correcciones, como verificar la valencia de los átomos, la validez de los enlaces, y otras propiedades estructurales. Sin embargo, si la molécula tiene una estructura no válida o inconsistente (por ejemplo, en tu caso un error con la valencia o la geometría), puede causar un error durante este proceso de sanitización.


```python
from rdkit import Chem

smiles = "[C@H]1(C=C(=O)=O)NC=C=C1"

molecula = Chem.MolFromSmiles(smiles, sanitize=True)
molecula
```

Cuando configuras `sanitize=False`, RDKit omite la verificación de estos problemas estructurales. Esto permite que la molécula sea creada aunque tenga una valencia no admitida o algún otro problema estructural, pero hay que tener en cuenta que la molécula podría no ser completamente válida desde un punto de vista químico.

```python
from rdkit import Chem

smiles = "[C@H]1(C=C(=O)=O)NC=C=C1"

molecula = Chem.MolFromSmiles(smiles, sanitize=False)
molecula
```

Si un SMILES está causando un error, tal vez sea útil revisar si la estructura es válida o probar sanitize=False si solo necesitas obtener la molécula sin importar si está completamente correcta.

```python
# Buscar los átomos que coinciden con la subestructura
hit_atomos = list(molecula.GetSubstructMatch(substructura))

# Generar una lista para guardar los índices de los átomos que coinciden con la subestructura
hit_enlaces = []
for enlace in substructura.GetBonds():
    aid1 = hit_atomos[enlace.GetBeginAtomIdx()]
    aid2 = hit_atomos[enlace.GetEndAtomIdx()]
    hit_enlaces.append(molecula.GetBondBetweenAtoms(aid1,aid2).GetIdx())

# Dibujar la molecula resaltando los átomos que se encuentran en la subestructura
d = rdMolDraw2D.MolDraw2DSVG(500, 500)
d = rdMolDraw2D.PrepareAndDrawMolecule(d, molecula, highlightAtoms=hit_atomos,
                                   highlightBonds=hit_enlaces)
molecula
```
