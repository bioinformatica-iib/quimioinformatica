## Variables

```python
x = 38
y = "Hola Mundo"

print(x)
print(y)
```

Por más que la variable contenga el mismo dato, son elementos diferentes

```python
z = 38
print(z)
```

Para saber el tipo de variable, puedo usar la función *type*

```python
print(type(x)) # x era 38
print(type(z)) # z era 38
print(type(y)) # y era "Hola Mundo"
```

### Reglas de Nombre de Variables

```python
mivariable = "Hola"
MIVARIABLE = "Hola2"

print(mivariable)
print(MIVARIABLE)
```

Los nombres de las variables no pueden comenzar con un número:

```python
2miVar = "Hola"
print(2miVar)
```

Los nombres de las variables no pueden contener espacios:

```python
mi var="Hola"
```

```python
mi_var="Hola" # asigno hola a mi_var
print(mi_var)
```

## Operaciones con variables

```python
x = 3
y = 4
print(x + y)
print(x - y)
print(x / y)
print(x * y)
print(x % y)
print(y % x)
print(x % 2)
print(y % 2)
```

## Operador Modulo: Ejemplo de uso para saber si un número es par o no

```python
7 % 2 == 1 # True --> impar
print(7 % 2 == 1)
print(25 % 5)
print(25 % 7)
24 % 2 == 0 # True --> Es par
print(24 % 2 == 0)
```

## Formas de concatenar strings

```python
x = "Hola"
y = " Mundo"
print(x + y)

x = "Hola"
y = "Mundo"
print(f"{x} {y}")

x = "Hola"
y = "Mundo"
print(x + " " + y)

# Pero con el +, si tengo una variable que es un número,
# la tengo que transformar a string con la función "str"
# para poder concatenerla a otro string
z = 5
print(x + str(z))
```

## Operadores lógicos

Permiten evaluar condiciones y dan como resultado un boolean True o False

```python
# Operadores Lógicos
x = True
y = False
print(x and y)
print(x or y)
print(not x)

# Otros operadores lógicos
print(3 > 4)
print(3 < 4)
print(3 >= 4)
print(3 <= 4)
print(3 == 4)
print(3 != 4)

x = "Hola"
y = "Mundo"
print(x == y)
print(x != y)
```

## Condicional if
Permiten crear secciones de
código que se van a ejecutar solo si se cumple
(o no se cumple) una condición.

Las condiciones dan como resultado un bool: True o False

```python
x = 8

x==8 # True
x == 7 # False
x != 8 # False

x < 10 # True
x > 10 # False

x = "Hola Mundo"
x == "hola mundo"
len(x) == 20 # False

"Mundo" in x # True

type(x) == str # True

if x == "Hola Mundo":
  print("Buen día!!")
else:
  print("Buenas noches!!")

print("Ya es mañana")
```

## Ciclo for
Permite iterar, de manera que cuando es necesario ejecutar el mismo código muchas veces no lo tenga que repetir muchísimas veces

```python
for i in [4,20,3,4]:
  print(i)
```

### range(start,stop,step)

La funcion range da una lista de números que empieza en el _start_ , avanza según el _step_ y termina en el _stop_ (pero no es un incluido)



```python
for i in range(1,10,1):
  print(i)
```

## Torta Mascarpone: Ejemplo de modificar una variable adentro de un loop

```python
remojar_en = "leche" # puede ser cafe o leche
espolvorear_con="cacao en polvo" # puede ser cacao en polvo o azucar impalpable
deseo_espolvorear = "No" # puede ser si o no
ingredientes = ["crema","queso mascarpone","vainillas"]

print(f"Mezclar el {ingredientes[0]} con el {ingredientes[1]}")

print("Preparar una fuente")

for i in range(1,4):
  print(f"Remojar {ingredientes[2]} en {remojar_en}")
  print(f"Crear la capa {i} de {ingredientes[2]}")
  print("Colocar la capa de mezcla")
  if remojar_en == "cafe":
    remojar_en = "vino"
  else:
    remojar_en = "cafe"

print("Enfriar en el refrigerador")

if deseo_espolvorear == "Si":
  print(f"Espolvorear con {espolvorear_con}")
```
